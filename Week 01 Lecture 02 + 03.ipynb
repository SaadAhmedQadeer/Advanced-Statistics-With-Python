## Part 1: NumPy - Numerical Computing Foundation

**Why Numpy:**
- NumPy is the foundation for numerical computing in Python
- It provides fast operations on arrays and matrices
- Essential for statistical calculations like mean, median, standard deviation
- Works with homogeneous data (all same type)

**Key Concept:** Arrays are faster than Python lists because they store data contiguously in memory

```python
# Import NumPy with standard alias 'np'
# Convention: Always use 'np' as the alias for NumPy
import numpy as np

# Creating arrays - the fundamental data structure in NumPy
# Arrays are faster and more memory-efficient than Python lists


# 1D array from a Python list
data = np.array([23, 45, 67, 89, 34, 56, 78, 90, 12, 45])
print("1D Array (student scores):")
print(data)
print(f"Type: {type(data)}\n")
```

```python
# Statistical Functions in NumPy
# These are the building blocks for statistical analysis

# np.mean() - calculates the arithmetic average
# Returns: sum of all elements / number of elements
mean_score = np.mean(data)
print(f"Mean (Average): {mean_score}")

# np.median() - finds the middle value when data is sorted
# Better than mean when data has outliers
median_score = np.median(data)
print(f"Median (Middle value): {median_score}")

# np.std() - standard deviation measures spread of data
# Shows how much data deviates from the mean
# Higher std = more spread out data
std_score = np.std(data)
print(f"Standard Deviation: {std_score:.2f}")

# np.var() - variance is the square of standard deviation
# Measures average squared deviation from mean
var_score = np.var(data)
print(f"Variance: {var_score:.2f}")

# np.min() and np.max() - find extreme values
print(f"Minimum score: {np.min(data)}")
print(f"Maximum score: {np.max(data)}")
```

```python
# Creating arrays with built-in functions
# These are useful for simulations and testing

# np.arange(start, stop, step) - creates evenly spaced values
# Similar to Python's range() but returns NumPy array
# stop value is EXCLUDED
numbers = np.arange(0, 10, 2)  # Start at 0, stop before 10, step by 2
print(f"Evenly spaced array: {numbers}")

# np.linspace(start, stop, num) - creates num evenly spaced points
# stop value is INCLUDED
# Useful for creating x-axis values for plotting
linear_space = np.linspace(0, 1, 5)  # 5 points from 0 to 1
print(f"Linear space: {linear_space}")

# np.random.randn(n) - generates n random values from standard normal distribution
# Mean = 0, Standard Deviation = 1
# Useful for simulating statistical data
np.random.seed(42)  # Set seed for reproducibility
random_data = np.random.randn(100)
print(f"Random normal data (first 10): {random_data[:10]}")
print(f"Mean of random data: {np.mean(random_data):.2f} (should be close to 0)")
print(f"Std of random data: {np.std(random_data):.2f} (should be close to 1)")
```

## Part 2: Pandas - Data Manipulation & Analysis

**Why**
- Pandas builds on NumPy but works with heterogeneous data (mixed types)
- Two main structures: Series (1D) and DataFrame (2D table)
- Think of DataFrame as an Excel spreadsheet in Python
- Essential for cleaning, transforming, and analyzing real-world data

**Key Concept:** DataFrames have labeled rows and columns, making data more intuitive to work with

```python
# Import Pandas with standard alias 'pd'
# Convention: Always use 'pd' as the alias for Pandas
import pandas as pd

# Creating a DataFrame - the most important Pandas structure
# Dictionary keys become column names
# Dictionary values become column data
data_dict = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Math': [85, 92, 78, 95, 88, 76],
    'Science': [88, 85, 90, 92, 85, 80],
    'English': [92, 88, 85, 90, 94, 88]
}

# pd.DataFrame() - creates a DataFrame from dictionary
# Each key-value pair becomes a column
df = pd.DataFrame(data_dict)

print("Student Grades DataFrame:")
print(df)
```

```python
# Essential DataFrame Exploration Functions
# ALWAYS use these when you first load data

# .head(n) - shows first n rows (default n=5)
# Quick way to preview your data
print("First 3 rows:")
print(df.head(3))
print()

# .tail(n) - shows last n rows (default n=5)
# Useful to check if data loaded completely
print("Last 2 rows:")
print(df.tail(2))
print()

# .info() - provides DataFrame summary
# Shows: number of rows, columns, data types, memory usage
# Critical for understanding your data structure
print("DataFrame Information:")
print(df.info())
print()

# .shape - tuple showing (rows, columns)
# Not a function, it's an attribute (no parentheses)
print(f"Shape (rows, columns): {df.shape}")

# .columns - shows all column names
print(f"Column names: {df.columns.tolist()}")
```

```python
# Statistical Functions on DataFrames
# These work column-wise by default

# .describe() - generates comprehensive statistical summary
# Shows: count, mean, std, min, quartiles, max
# Only works on numeric columns
print("Statistical Summary:")
print(df.describe())
print()

# .mean() - calculates mean for each numeric column
# Returns a Series with column names as index
print("Mean scores per subject:")
print(df.mean(numeric_only=True))  # numeric_only=True skips non-numeric columns
print()

# .median() - calculates median for each numeric column
print("Median scores per subject:")
print(df.median(numeric_only=True))
print()

# .std() - standard deviation for each column
print("Standard deviation per subject:")
print(df.std(numeric_only=True))
```

```python
# Selecting and Filtering Data
# These are the most common operations in data analysis

# Selecting a single column - returns a Series
# Two ways: df['column'] or df.column (prefer first method)
math_scores = df['Math']
print("Math scores only:")
print(math_scores)
print(f"Type: {type(math_scores)}\n")

# Selecting multiple columns - returns a DataFrame
# Pass a list of column names inside square brackets
subset = df[['Student', 'Math']]
print("Student names and Math scores:")
print(subset)
print()

# Filtering rows based on conditions
# Creates a boolean mask, then uses it to filter
# Boolean indexing: df[condition]
high_math = df[df['Math'] > 85]  # Students with Math score > 85
print("Students with Math > 85:")
print(high_math)
print()

# Multiple conditions
# Use & for AND, | for OR
# Each condition must be in parentheses
high_performers = df[(df['Math'] > 85) & (df['Science'] > 85)]
print("Students with both Math AND Science > 85:")
print(high_performers)
```

```python
# Adding new columns
# Just assign to a new column name

# Calculate average across subjects for each student
# axis=1 means calculate across columns (row-wise)
# axis=0 would calculate down columns (column-wise)
df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)

print("DataFrame with Average column:")
print(df)
print()

# .sort_values() - sort DataFrame by column values
# by: column name to sort by
# ascending: True for low to high, False for high to low
df_sorted = df.sort_values(by='Average', ascending=False)
print("Students ranked by average score:")
print(df_sorted)
```

## Part 3: Matplotlib - Basic Visualization


- Matplotlib is the foundation of Python plotting
- pyplot module provides MATLAB-like interface
- Essential for creating statistical plots: histograms, scatter plots, line graphs
- Highly customizable but requires more code

**Key Concept:** Every plot has a Figure (container) and Axes (actual plot area)

```python
# Import matplotlib.pyplot with standard alias 'plt'
# Convention: Always use 'plt' as the alias
import matplotlib.pyplot as plt

# %matplotlib inline - magic command for Jupyter/Colab
# Makes plots appear directly in the notebook
%matplotlib inline
```

```python
# Histogram - shows distribution of continuous data
# Essential for understanding data spread and shape

# Generate sample data for demonstration
np.random.seed(42)
exam_scores = np.random.normal(75, 10, 100)  # 100 scores, mean=75, std=10

# plt.figure(figsize=(width, height)) - creates a new figure
# figsize is in inches, tuple of (width, height)
# Standard size is (10, 6)
plt.figure(figsize=(10, 6))

# plt.hist() - creates histogram
# x: data to plot
# bins: number of bars (more bins = more detail)
# edgecolor: color of bar borders (makes bars distinct)
# alpha: transparency (0=transparent, 1=opaque)
plt.hist(exam_scores, bins=15, edgecolor='black', alpha=0.7, color='skyblue')

# plt.title() - adds title to plot
# fontsize: size of title text
plt.title('Distribution of Exam Scores', fontsize=14, fontweight='bold')

# plt.xlabel() - labels x-axis
plt.xlabel('Score', fontsize=12)

# plt.ylabel() - labels y-axis
plt.ylabel('Frequency', fontsize=12)

# plt.grid() - adds gridlines
# alpha: transparency of gridlines
# Makes reading values easier
plt.grid(alpha=0.3)

# plt.show() - displays the plot
# Always call this at the end
plt.show()
```

```python
# Scatter Plot - shows relationship between two variables
# Used to detect correlation and patterns

# Create sample data: study hours vs exam scores
study_hours = np.random.uniform(1, 10, 50)  # 50 random values between 1 and 10
exam_scores = 45 + 5 * study_hours + np.random.normal(0, 5, 50)  # Linear relationship + noise

plt.figure(figsize=(10, 6))

# plt.scatter() - creates scatter plot
# x: x-axis data (independent variable)
# y: y-axis data (dependent variable)
# c: color of points
# s: size of points
# alpha: transparency
plt.scatter(study_hours, exam_scores, c='darkblue', s=50, alpha=0.6)

plt.title('Relationship Between Study Hours and Exam Scores', fontsize=14, fontweight='bold')
plt.xlabel('Study Hours', fontsize=12)
plt.ylabel('Exam Score', fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Line Plot - shows trends over time or continuous data
# Common for time series and trend analysis

# Sample data: test scores over 10 weeks
weeks = np.arange(1, 11)  # Week 1 to 10
scores = [65, 68, 72, 75, 78, 82, 85, 87, 90, 92]  # Improving trend

plt.figure(figsize=(10, 6))

# plt.plot() - creates line plot
# x: x-axis values
# y: y-axis values
# marker: shape of data points ('o'=circle, 's'=square, '^'=triangle)
# linestyle: line style ('-'=solid, '--'=dashed, ':'=dotted)
# linewidth: thickness of line
# color: line color
plt.plot(weeks, scores, marker='o', linestyle='-', linewidth=2,
         color='green', markersize=8, label='Student Progress')

plt.title('Student Progress Over 10 Weeks', fontsize=14, fontweight='bold')
plt.xlabel('Week Number', fontsize=12)
plt.ylabel('Test Score', fontsize=12)

# plt.legend() - shows legend for labeled plots
# loc: location of legend ('best', 'upper right', 'lower left', etc.)
plt.legend(loc='best')

plt.grid(alpha=0.3)
plt.show()
```

## Part 4: Seaborn - Statistical Visualization


- Seaborn builds on Matplotlib with better defaults and statistical focus
- Creates beautiful plots with less code
- Designed specifically for statistical analysis
- Integrates perfectly with Pandas DataFrames

**Key Concept:** Seaborn makes complex statistical plots easy and aesthetically pleasing

```python
import numpy as np
import pandas as pd
```

```python
# Import Seaborn with standard alias 'sns'
# Convention: Always use 'sns' as the alias
import seaborn as sns

# sns.set_style() - sets the aesthetic style of plots
# Options: 'darkgrid', 'whitegrid', 'dark', 'white', 'ticks'
# 'whitegrid' is clean and professional for presentations
sns.set_style('whitegrid')

# Create a comprehensive dataset for demonstration
np.random.seed(42)
sample_data = pd.DataFrame({
    'Math': np.random.normal(75, 12, 100),
    'Science': np.random.normal(78, 10, 100),
    'English': np.random.normal(72, 15, 100),
    'Gender': np.random.choice(['Male', 'Female'], 100),
    'Class': np.random.choice(['A', 'B', 'C'], 100)
})
```

```python
# Box Plot - shows distribution quartiles and outliers
# Essential for comparing distributions across categories

plt.figure(figsize=(10, 6))

# sns.boxplot() - creates box plot
# data: DataFrame to use
# x: categorical variable for x-axis
# y: continuous variable for y-axis
# palette: color scheme ('Set1', 'Set2', 'pastel', etc.)
# Box shows: 25th percentile, median (50th), 75th percentile
# Whiskers: extend to 1.5 * IQR
# Points beyond whiskers: outliers
sns.boxplot(data=sample_data, x='Class', y='Math', palette='Set2')

plt.title('Math Score Distribution by Class', fontsize=14, fontweight='bold')
plt.xlabel('Class', fontsize=12)
plt.ylabel('Math Score', fontsize=12)
plt.show()

# Teaching point: Explain how to read a box plot
# Bottom of box = 25th percentile
# Line in box = median (50th percentile)
# Top of box = 75th percentile
# Box height = Interquartile Range (IQR)
```

```python
# Violin Plot - combines box plot with kernel density
# Shows full distribution shape

plt.figure(figsize=(10, 6))

# sns.violinplot() - creates violin plot
# Wider sections = more data points
# Shows distribution shape better than box plot
# inner='box': adds mini box plot inside
# split: if True and hue is used, splits violins for comparison
sns.violinplot(data=sample_data, x='Gender', y='Science', palette='muted', inner='box')

plt.title('Science Score Distribution by Gender', fontsize=14, fontweight='bold')
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Science Score', fontsize=12)
plt.show()
```

```python
# Heatmap - visualizes correlation matrix
# Critical for understanding relationships between variables

plt.figure(figsize=(8, 6))

# Calculate correlation matrix
# .corr() - computes pairwise correlation of columns
# Values range from -1 (negative correlation) to +1 (positive correlation)
# 0 = no correlation
correlation_matrix = sample_data[['Math', 'Science', 'English']].corr()

# sns.heatmap() - creates heatmap
# annot=True: shows correlation values in cells
# fmt='.2f': formats numbers to 2 decimal places
# cmap: color map ('coolwarm' shows positive/negative clearly)
# cbar: whether to show color bar
# square=True: makes cells square-shaped
# linewidths: width of lines between cells
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm',
            cbar=True, square=True, linewidths=1)

plt.title('Correlation Matrix of Subject Scores', fontsize=14, fontweight='bold')
plt.show()

# Teaching point: Interpret correlation values
# Close to 1: strong positive correlation (both increase together)
# Close to -1: strong negative correlation (one increases, other decreases)
# Close to 0: little to no correlation
```

```python
# Pair Plot - shows relationships between all numeric variables
# Combines scatter plots and histograms
# Excellent for exploratory data analysis

# sns.pairplot() - creates matrix of plots
# data: DataFrame to plot
# hue: color points by category
# diag_kind: type of plot on diagonal ('hist' or 'kde')
# palette: color scheme for hue categories
# Diagonal: distribution of each variable
# Off-diagonal: scatter plots showing relationships
pair_plot = sns.pairplot(sample_data[['Math', 'Science', 'English', 'Gender']],
                         hue='Gender', diag_kind='hist', palette='husl')

pair_plot.fig.suptitle('Pairwise Relationships Between Subjects', y=1.02, fontsize=14, fontweight='bold')
plt.show()

# Teaching point: How to read pairplot
# Each scatter plot shows relationship between two variables
# Look for patterns: linear, curved, clustered, or random
# Colors separate by categories (Gender here)
```

```python
# Count Plot - bar chart for categorical data
# Shows frequency of each category

plt.figure(figsize=(10, 6))

# sns.countplot() - creates count plot
# x: categorical variable to count
# hue: split bars by another category
# data: DataFrame containing the data
# palette: color scheme
sns.countplot(data=sample_data, x='Class', hue='Gender', palette='Set1')

plt.title('Student Distribution by Class and Gender', fontsize=14, fontweight='bold')
plt.xlabel('Class', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Gender')
plt.show()
```
